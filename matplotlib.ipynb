```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=x*x
z=x+y
```

```python
plt.plot(x,x,label="xx")
plt.plot(x,y,label="xy")
plt.plot(x,z,label="xz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("practic")
plt.legend()
```

# subplot
(rows,columns,no of graph)

legend(loc=1) loc means location it start from 1 to 10

```python
plt.subplot(3,2,1)
plt.plot(x,x, label="xx")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=1)
plt.subplot(3,2,2)
plt.plot(x,y, label="xy")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=4)
plt.subplot(3,2,3)
plt.plot(x,z, label="xz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=5)
plt.subplot(3,2,4)
plt.plot(y,z,label="yz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=8)
plt.subplot(3,2,5)
plt.plot(z,x,label="zx")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=2)
plt.subplot(3,2,6)
plt.plot(z,y,label="zy")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=9)
plt.tight_layout() #give space between graphs
```

```python
#to control figure 
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,1,1]) #start of x ,start of y,end of x,end of y
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('new way to add title and label')
axes.plot(x,y)
```

```python
#multiple graph in a singel graph
fig=plt.figure()
x1=fig.add_axes([0.1,0.1,1,1])
x1.plot(x,y)
x1.set_xlabel('x')
x1.set_ylabel('y')

x2=fig.add_axes([0.4,0.3,0.3,0.3])
x2.plot(x,x)
x2.set_xlabel('x')
x2.set_ylabel('x')
x3=fig.add_axes([0.7,0.3,0.3,0.3])
x3.plot(z,x)
x3.set_xlabel('z')
x3.set_ylabel('x')
x4=fig.add_axes([0.1,0.3 ,0.3,0.3])
x4.plot(y,x)
x4.set_xlabel('y')
x4.set_ylabel('x')
```

```python
#subplot with mutliple grph

fig , axes=plt.subplots(1,2)

axes[0].plot(x,y)

axes[0].set_xlabel('y')
axes[0].set_ylabel('x')
axes[1].plot(y,x)
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
plt.tight_layout()
```

```python
fig , axes=plt.subplots(2,2)

axes[(0,0)].plot(x,y)
axes[0,1].plot(y,x)
axes[0,1].plot(y,x)
axes[0,1].plot(y,x)
plt.tight_layout()
```

```python
#figure size

fig=plt.figure(figsize=(5,4))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(2,1, figsize=(15,15))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig=plt.figure(figsize=(5,4))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='r', linewidth=20,alpha=0.5,linestyle="--") #styling
```

```python
#marker
fig=plt.figure(figsize=(5,4))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y, marker='o',markersize=10,markerfacecolor='red',markeredgecolor='g',markeredgewidth=4)
```

```python
#setting limit
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlim([0,10])
axes.set_ylim([0,60])
```
